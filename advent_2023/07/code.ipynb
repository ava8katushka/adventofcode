```python
from functools import cmp_to_key
import os
from collections import Counter, defaultdict
file_path = os.path.join(os.getcwd(), "input.txt")
score = 0

def extract_numbers(line):
        l = filter(lambda x: len(x), line.split(" "))
        return list(map(lambda x: int(x.strip()), l))

def get_type(hand):
    c = Counter(hand)
    jokers = c["J"]
    del c["J"]
    order = sorted(c.values(), reverse=True)
    first = order[0] if len(order) > 0 else 0
    second = order[1] if len(order) > 1 else 0
    if first + jokers == 5:
         return 0
    if first + jokers == 4:
         return -1
    if first + jokers == 3 and second == 2:
         return -2
    if first + jokers == 3:
         return -3
    if first == 2 and second == 2:
         return -4
    if first + jokers == 2:
         return -5
    return -6

def compare(hand_a, hand_b):
     type_a = get_type(hand_a)
     type_b = get_type(hand_b)
     if type_a < type_b or type_a > type_b:
          return type_a - type_b
     if hand_a == hand_b:
          return 0
     order = "A,K,Q,T,9,8,7,6,5,4,3,2,J".split(",")
     i = 0
     while hand_a[i] == hand_b[i]:
          i += 1
     return  order.index(hand_b[i]) - order.index(hand_a[i])

def pair_compair(a, b):
    return compare(a[0], b[0])
     
        

with open(file_path, encoding ="utf-8") as f:
    file = f.read().strip().split("\n")
    hand_bid = []
    for line in file:
        arr = line.split(" ")
        hand_bid.append([arr[0], int(arr[1])])
    hand_bid = sorted(hand_bid, key=cmp_to_key(pair_compair), reverse=True)
    n = len(hand_bid)
    for i, [hand, bid] in enumerate(hand_bid):
         score += (n - i) * bid
         if i > 900:
              print(hand, n - i, bid)
         
    print(get_type("JJJJJ"))
    
    

    
# 250736593 -> too low
# 251037509
print(score)
        
                  
             
    

        







        

```
