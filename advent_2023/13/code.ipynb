```python
from functools import cmp_to_key
import os
from collections import Counter, defaultdict, deque
import re
import sys
from copy import deepcopy

file_path = os.path.join(os.getcwd(), "input.txt")
score = 0

def stringify(arr):
      lines = ["".join(arr[i]) for i in range(len(arr))]
      return "\n".join(lines)

def dist(a, b):
    diff = 0
    for i in range(len(a)):
          if a[i] != b[i]:
                diff += 1
    return diff


def is_true_reflection(arr, i):
      stack = []
      res = 0
      for j in range(i):
            stack.append(arr[j])
      for j in range(i, len(arr)):
            if len(stack):
                  if stack[-1] != arr[j]:
                        res += dist(stack[-1], arr[j])
                        stack.pop()
                  else:
                        stack.pop()
      return res


def find_horizontal(arr):
      for i in range(1, len(arr)):
            if is_true_reflection(arr, i) == 1:
                  return i
      return -1





with open(file_path, encoding ="utf-8") as f:
      file = f.read().strip().split("\n\n")
      score = 0
      for num, pattern in enumerate(file):
            arr = pattern.split("\n")
            h = find_horizontal(arr)
            if h != -1:
                  print(num, h)
                  score += h * 100
                  continue
            r = list(zip(*arr[::-1]))
            rotated = ["".join(r[i]) for i in range(len(r))]
            v = find_horizontal(rotated)
            print(num, v)
            score += v
      print(score)
      # 35727 too low
      # 39939
      # 113200 too high

      



      
```
