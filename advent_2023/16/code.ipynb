```python
from functools import cmp_to_key
import os
from collections import Counter, defaultdict, deque
import re
import sys
from copy import deepcopy

file_path = os.path.join(os.getcwd(), "input.txt")
score = 0

def extract_numbers(line):
        l = filter(lambda x: len(x), line.split(","))
        return tuple(map(lambda x: int(x.strip()), l))

def get_arrangments(cur, line, i, res):
      if i == len(line):
             res.append(cur)
             return
      if line[i] == "?":
             get_arrangments(cur + ".", line, i + 1, res)
             get_arrangments(cur + "#", line, i + 1, res)
      else:
             get_arrangments(cur + line[i], line, i + 1, res)

def get_nums(line):
      cur = 0
      res = []
      for char in line:
            if char == "#":
                  cur += 1
            elif cur > 0:
                  res.append(cur)
                  cur = 0
      if cur > 0:
            res.append(cur)
      return tuple(res)
                     
      

with open(file_path, encoding ="utf-8") as f:
      file = f.read().strip().split("\n")
      for line in file:
            springs, nums = line.split(" ")
            nums = extract_numbers(nums)
            arrs = []
            get_arrangments("", springs, 0, arrs)
            var = 0
            for arr in arrs:
                  #print(arr, get_nums(arr))
                  if get_nums(arr) == nums:
                        var += 1
            print(springs, nums, var)
            score += var
      print(score)



      
```
