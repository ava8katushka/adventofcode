```python
from functools import cmp_to_key
import os
from collections import Counter, defaultdict, deque
import re
import sys
from copy import deepcopy

file_path = os.path.join(os.getcwd(), "input.txt")
score = 0

def extract_numbers(line):
        l = filter(lambda x: len(x), line.split(" "))
        return list(map(lambda x: int(x.strip()), l))

def is_empty(line):
    for char in line:
          if char == "#":
                return False
    return True
     


with open(file_path, encoding ="utf-8") as f:
      file = f.read().strip().split("\n")
      space = [list(line) for line in file]
      empty_lines = []
      cur = 0
      for line in space:
          if is_empty(line):
                cur += 1
          empty_lines.append(cur)
      #00011112222
      cur = 0 
      empty_cols = []
      for line in zip(*space[::-1]):
          if is_empty(line):
                cur += 1
          empty_cols.append(cur)
      galaxy = []
      for i, line in enumerate(space):
            for j, char in enumerate(line):
                  if char == "#":
                        galaxy.append((i, j))
      coeff = 1000000
      for i in range(len(galaxy)):
           for j in range(i + 1, len(galaxy)):
                a, b = galaxy[i], galaxy[j]
                x = abs(b[0] - a[0]) # (i, j)
                lines = abs(empty_lines[b[0]] - empty_lines[a[0]])
                y = abs(b[1] - a[1])
                cols = abs(empty_cols[b[1]] - empty_cols[a[1]])
                shift =  x + y + (lines + cols) * (coeff - 1)
                #print(a, b, "=", shift, "of", x, y, "lines", lines, "cols", cols)
                score += shift
                
      print(score)
                     


      
















             


    
         
         

    
        
                  
             
    

        

#     Between galaxy 1 and galaxy 7: 15
#     Between galaxy 3 and galaxy 6: 17
#     Between galaxy 8 and galaxy 9: 5

# 3 6 = 16 of 4 9 lines 0 cols 3
# 3 7 = 16 of 6 7 lines 1 cols 2
# 8 9 = 6 of 0 4 lines 1 cols 1







        

```
