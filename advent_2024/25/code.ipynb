```python
from functools import cmp_to_key
import math
import os
from collections import Counter, defaultdict, deque
import random
import re
import sys
from copy import deepcopy
from fractions import Fraction
import networkx

file_path = os.path.join(os.getcwd(), "input.txt")
score = 0

class Node:
      def __init__(self, id):
            self.id = id
            self.parent = self
            self.size = 1

      
def find_parent(node):
      if node.parent is node:
            return node
      node.parent = find_parent(node.parent)
      return node.parent

def union(a, b):
      node_a = find_parent(a)
      node_b = find_parent(b)
      if node_a is node_b:
            return node_a
      if node_a.size < node_b.size:
            node_a, node_b = node_b, node_a
      node_b.parent = node_a
      node_a.size += node_b.size
      return node_a
      



with open(file_path, encoding ="utf-8") as f:
      file = f.read().strip().split("\n")
      min_size = math.inf
      it = 0
      while min_size > 3:
            it += 1
            graph = defaultdict(list)
            edges = []
            for line in file:
                  a, other = line.split(":")
                  other = other.strip().split(" ")
                  for b in other:
                        graph[a].append(b)
                        graph[b].append(a)
                        edges.append((a, b))
            nodes = {}
            separate_sets = set()
            for key in graph:
                  nodes[key] = Node(key)
                  separate_sets.add(key)
            random.shuffle(edges)
            i = 0
            # kargers algorithm
                        
            while len(separate_sets) > 2 and i < len(edges):
                  i += 1
                  a,b = edges[i]
                  pa, pb =  find_parent(nodes[a]), find_parent(nodes[b])
                  if pa == pb:
                        continue
                  together = union(nodes[a], nodes[b])
                  separate_sets.remove(pa.id)
                  separate_sets.remove(pb.id)
                  separate_sets.add(together.id)
                  
            pa_a, pa_b = separate_sets
            a_edge, b_edge = [], []
            for node in graph:
                  pa_node = find_parent(nodes[node])
                  if pa_node.id == pa_a:
                        for c in graph[node]:
                              if find_parent(nodes[c]).id != pa_a:
                                    a_edge.append(node)
                  elif pa_node.id == pa_b:
                        for c in graph[node]:
                              if find_parent(nodes[c]).id != pa_b:
                                    b_edge.append(node)
                  else:
                        print("Ooops!")

            assert(len(a_edge) == len (b_edge))
            if (new_min := len(a_edge)) < min_size:
                  min_size = new_min
                  print(f"{it}) {min_size}")
            score = nodes[pa_a].size * nodes[pa_b].size
      print(score)



                  

            
      

      



      
```
